```python
import numpy as np
import pandas as pd
```

```python
given_data=pd.read_csv(r"C:\Users\hgaikw\Downloads\DAX-Freshers-Training-main\DAX-Freshers-Training-main\healthcare-dataset-stroke-data.csv")
```

```python
given_data.head()
```

```python
given_data.drop(['id'], axis=1, inplace=True)
given_data.head()
```

```python
given_data.isnull()
```

```python
given_data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

colList = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']

scaler = StandardScaler()

scaledData = given_data[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']]
scaledData = pd.DataFrame(scaler.fit_transform(scaledData), columns=colList)
print(scaledData)
```

```python
#Adding scaled data to given data
given_data['age'] = scaledData['age'] 
given_data['hypertension'] = scaledData['hypertension']
given_data['heart_disease'] = scaledData['heart_disease']
given_data['avg_glucose_level'] = scaledData['avg_glucose_level']
given_data['bmi'] = scaledData['bmi']
```

```python
given_data.head()
```

```python
category_data = given_data.select_dtypes(include=['object']).copy()
category_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

colList = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

for i in colList:
    encoder.fit(given_data[i])
    given_data[i] = encoder.transform(given_data[i])

given_data.head()
```
